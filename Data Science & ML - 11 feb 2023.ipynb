```python
import pandas as pd
```

```python
df = pd.read_csv('medical_mimic.csv')
```

```python
df.head()
```

```python
# unique values of column

df.columns
```

```python
for column in df.columns:
    uniq_counts = len(df[column].unique())
    print(column,'=>',uniq_counts)
    
    # Drop column if unique value count is 1     
    if uniq_counts == 1:
        print('Dropping ',column)
        df = df.drop(columns=[column])
```
